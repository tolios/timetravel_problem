```python
class agent():
    #agent class performing the decisions and trading...
    
    def __init__(self, patience = 500, margin = 4.0, final_day = -1):
        self.patience = patience
        self.margin = margin
        self.wallet = lib.wallet()
        self.memory = dict()
        self.final_day = final_day
        
    def act(self, stocks, dates, day, day_index):
 
        daily_stocks = stocks[day]
        
    
        actions = self.death_policy(day)
        if not actions:
            actions = self.coughing_policy(stocks, dates, day_index)
        if not actions:
            actions = self.infection_policy(daily_stocks, day)
        if not actions:
            actions = self.fever_policy(daily_stocks, day)
        if not actions:
            actions = self.incubation_policy(daily_stocks, day)
        return actions
        
    def death_policy(self, day):
        '''
        When the day comes, sell everything you own!
        '''
        actions = []
        if day == self.final_day:
            for company in self.wallet.own_stocks:
                stocks = self.wallet.own_stocks[company]
                actions.append(self.execute(day, 'sell-high', company, stocks))
        return actions
    
    def coughing_policy(self, stocks, dates, day_index):
        #memory of best day sell!
        #selling policy
        day = dates[day_index]
        actions = []            
    
        for company in self.wallet.own_stocks.copy():
            if company in self.memory:
                if self.memory[company] == day:
                    if day_index == 13962 or day_index == 13961 or day_index == 13963:
                        print(stocks[day])
                        print(day)
                        print(company)
                        print(self.wallet.own_stocks[company])
                    actions.append(self.execute(day, 'sell-high',
                                    company, self.wallet.own_stocks[company]))
                    del self.memory[company]
            else:
                if self.wallet.own_stocks[company] > 1:
                    self.premonition(company, stocks, dates, day_index)
        return actions
        
    def fever_policy(self, daily_stocks, day):
        '''
        Buys as many stocks as possible
        '''
        actions = []
        #buy stocks if possible
        for company in self.wallet.own_stocks.copy():
            if company in daily_stocks:
                stocks = self.wallet.own_stocks[company]
                _, _, low, _, vol = daily_stocks[company]
                for able_stocks in range(stocks+1, 0, -1):
                    if able_stocks*low < self.wallet.balance[-1] + self.wallet.balance_change:
                        actions.append(self.execute(day, 'buy-low', company, able_stocks))
                        break
        return actions
        
    def infection_policy(self, daily_stocks, day):
        #infect new company!
        actions = []
        infection = False
        min_ = 1e6
        for company in daily_stocks:
            if company not in self.wallet.own_stocks:
                _, _, low, _, _ = daily_stocks[company]
                if low < min_:
                    infection = True
                    company_infected = company
                    min_ = low
                    
        if infection and self.wallet.balance[-1] > min_:
            #perform infection!
            actions.append(self.execute(day, 'buy-low', company_infected, 1))
            
        return actions
    
    def incubation_policy(self, daily_stocks, day):
        actions = []
        #this policy activates when you simply want
        #to keep current stocks, and generate small profits
        #intra day trading!
        print('ho')
        for company in daily_stocks:
            if company in self.wallet.own_stocks:
                #if we have that company in our stocks!
                #want to buy only when profits are acceptable
                open_, high, low, close, vol = daily_stocks[company]
                #find biggest profit action per day
                profit1 = float(high) - float(close)
                profit2 = float(open_) - float(low)
                #if not positive, we shouldn't trade!
                if profit1 >= profit2 and profit1 > self.margin:
                    print('he')
                    #sell high, buy close
                    stock_num = self.wallet.own_stocks[company]
                    actions.append(self.execute(day, 'sell-high', company, stock_num))
                    actions.append(self.execute(day, 'buy-close', company, stock_num))
                if profit2 > profit1 and profit2 > self.margin:
                    print('he')
                    #sell open, buy low
                    stock_num = self.wallet.own_stocks[company]
                    actions.append(self.execute(day, 'sell-open', company, stock_num))
                    actions.append(self.execute(day, 'buy-low', company, stock_num))
                
        return actions
        
    def execute(self, date, action, name, stocks):
        #basically executes the wallet action
        valid = self.wallet.execute(date, action, name, stocks)
        return str(date)[:-9]+' '+action+' '+name+' '+str(stocks)
    
    def premonition(self, company, stocks, dates, day_index):
        '''
        Finds best day to sell, inside a window
        of days and years!
        '''
        max_sell = -1
        profit_day = None
        #find best sell day within time window...
        for date in dates[day_index: day_index+self.patience]:
            daily_stock = stocks[date]
            if company in daily_stock:
                _, high, _, _, _ = daily_stock[company]
                if high >= max_sell:
                    max_sell = high
                    profit_day = date
        #memory entry!
        if profit_day != dates[day_index]:
            self.memory[company] = profit_day
        else:
            self.memory[company] = dates[day_index+1]
    
    def update_wallet(self, day):
        self.wallet.update(day)
```

```python
bond_james = agent(patience = 700, margin = 1.0)
bond_james.wallet.portofolio = [10.]
N = 0
mems = []
'''
VERY IMPORTANT
needs premonition for buying!!!

'''
start = 3000
finish = 5000
print(finish - start)

for day_index, day in tqdm(enumerate(dates[start:finish]), desc = 'Years passing by...'):
    
    action = bond_james.act(stock_dataset, dates, day, day_index+start)
    if action:
        N += len(action)
        mems.extend(action)
    bond_james.update_wallet(day)

print(N)
print(mems)
#13962 error with start 900, in coughing policy cant convert float a series
```
